# Challenge Data Enginner


Importar datos desde el archivo .json que se encuentra dentro del archivo .zip

```python
from zipfile import ZipFile
from time import time
import os

#Definimos la función para extraer los tweets del .zip
def extract_tweets(target_file: str, zip_file_path: str) -> None:
    """
    Extrae tweets de un archivo zip si el archivo de destino no existe.

    Variables:
        file_path (str): Ruta al archivo de destino.
        zip_file_path (str): Ruta al archivo zip que contiene los tweets.
    """
    if not os.path.exists(target_file):
        with ZipFile(zip_file_path, 'r') as zip_object:
            zip_object.extractall('.')          
    

#Configurar las rutas de entrada y de salida
file_path = "farmers-protest-tweets-2021-2-4.json"
zip_file_path = "../src/tweets.json.zip"

#Llamar a la función para extraer los tweets
extract_tweets(file_path, zip_file_path)

#Imprimir un mensaje si se extrajeron los tweets de forma exitosa
print(f"Los tweets se extrajeron de la ruta: {zip_file_path} en el archivo: {file_path}")

        
```
